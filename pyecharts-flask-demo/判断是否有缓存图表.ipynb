```python
import os

# 指定文件夹路径和文件名
folder_path = "charts"
file_name = "上海_产品经理_2023-12-13_map.html"

# 构建完整的文件路径
file_path = os.path.join(folder_path, file_name)

# 检查文件是否存在
if os.path.exists(file_path):
    print("文件存在")
    with open(file_path,"r") as file:
        content = file.read()
        print(content)
else:
    print("文件不存在")
```
